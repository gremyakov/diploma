```python
import numpy as np
p = 5
n = 1
q = p^n
F = FiniteField(q,'a')
a = F.gen()

if n > 1:
    Set = range(1,q)
    Set = [a^j for j in Set]
    Set = [0] + Set
else: 
    Set = range(1,q)
    Set = [(a+a+a)^j for j in Set]
    Set = [0] + Set
print vector(Set)
#G = SymmetricGroup([Set[1],Set[2],Set[0],Set[3]])
G = SymmetricGroup(Set)
```

```python
def Matrica(F,p,n,Set):
    M = matrix(F, q, q)
    
    M[0,0] = -1

    for i in range(1,q-1):
        k = q-(i+1)
        for j in range(1,q):
            M[i,j] = (Set[j])^(k)

    for i in range(q):
        M[q-1,i] = 1
    
    M = -M
    
    return M

print Set
M = Matrica(F,p,n,Set)
print M 
```

```python
def orbit_of(vctr):
    orbit = []
    for g in G:
        tmp = []
        for item in vctr:
            tmp.append( g.dict()[item] )
        orbit.append(tmp)
    return orbit
        
def orbit_first_type(vctr):
    orbit = []
    for g in G:
        tmp = []
        for item in vctr:
            tmp.append( g.dict()[item] )
        orbit.append( g.inverse()(tmp) )
    return orbit

def remove_repetitions(lst):
    tmp = []
    for i in lst:
        if i not in tmp:
            tmp.append(i)
    return tmp

def number_of_different_values(lst):
    return len( set(lst) )


def get_vector(m, qq=q, lst=Set):
    if m>=qq**qq: 
        raise Exception('your argument of function is ' 
                        + str(m) 
                        +' but it must be less then '
                        + str(qq**qq))
    tmp = []
    while m!=0:
        remr = m%qq
        m = (m-remr)/qq
        tmp.insert(0,lst[remr])
        
    for i in xrange(qq-len(tmp)):
        tmp.insert(0,lst[0])
    return tmp


def nz_pol(pol):
    i=0
    for c in pol:
        if c!=0:
            i+=1
    return i

def key_nz(lst):
        return lst[0]
    
def key_nz_class(lst):
        return lst[0][0]
```

```python
classes1 = []
in_classes1 = []
vctr1 = []
orbit1 = []
d = q^q
c = 0
for i in xrange(q^q):
    c = 0
    vctr1 = get_vector(i)
    for k in xrange(len(in_classes1)):
        if vctr1 == in_classes1[k]:
            c = 1
            break
    if c == 1:
        continue
    else:
        orbit1 = remove_repetitions(orbit_first_type(vctr1)) 
        for j in xrange(len(orbit1)):
            in_classes1.append(orbit1[j])
        classes1.append(orbit1)
        
        
        if len(in_classes1) == d:
            break
            
for cls in classes1:
    for i in xrange(len(cls)):
        pol = list(M*vector(cls[i]))
        cls[i] = [nz_pol(pol), cls[i], list(M*vector(cls[i]))]
        
for cls in classes1:
    cls.sort(key = key_nz)
classes1.sort(key = key_nz_class, reverse = True)
```
